```python
class Book:
    def __init__(self, title, author, price, quantity):
        self.title = title
        self.author = author
        self.price = price
        self.quantity = quantity
        self.available = quantity  # Tracks available copies

    def borrow(self):
        if self.available > 0:
            self.available -= 1
            return True
        else:
            return False

    def return_book(self):
        self.available += 1

    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nPrice: {self.price}\nAvailable: {self.available}"
```

```python
class Member:
    def __init__(self, name, ID, plan, contributions=0, borrowed_books=None):
        self.name = name
        self.ID = ID
        self.plan=plan
        self.contributions=contributions
        self.borrowed_books = []


    def borrow_book(self, library, book_title):
        book = library.find_book(book_title)
        if book and book.borrow():
            self.borrowed_books.append(book)
            return True,book
        else:
            return False

    def return_book(self, library, book_title):
        book = library.find_book(book_title)
        if book in self.borrowed_books:
            book.return_book()
            self.borrowed_books.remove(book)
            return True
        else:
            return False
        
        

    def __str__(self):
        return f"Name: {self.name}\nMember ID: {self.ID}\nMembership Plan: {self.plan}\nBorrowed Books: {', '.join(str(book) for book in self.borrowed_books)}\nContributions: {self.contributions}"
```

```python
class Library:
    def __init__(self, books=[], members=[]):
        self.books = books
        self.members = members
    
    def add_member(self, member):
        self.members.append(member)

    def add_book(self, book):
        self.books.append(book)

    def find_book(self, book_title):
        for book in self.books:
            if book.title == book_title:
                return book
        return False
    
    def find_member(self,member_id):
        for member in self.members:
            if member.ID==member_id:
                return member
        print("Please check your ID. If not a member, Please register.")

    def print_available_books(self):
        print("Available Books:")
        for book in self.books:
            if int(book.available) > 0:
                print(f"- {book}")

    def print_members(self):
        print("Members:")
        for member in self.members:
            print(f"- {member}")

    def __str__(self):
        return f"Library with {len(self.books)} books and {len(self.members)} members"
```
